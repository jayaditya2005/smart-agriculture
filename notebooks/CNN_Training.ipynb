```python
# 2. Import Libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import os
import matplotlib.pyplot as plt
```

```python
# 3. Prepare Image Data Generators (20% for validation)
base_dir = 'dataset_folder'

datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_generator = datagen.flow_from_directory(
    base_dir, 
    target_size=(150, 150), 
    batch_size=32, 
    class_mode='categorical', 
    subset='training'
)

val_generator = datagen.flow_from_directory(
    base_dir, 
    target_size=(150, 150), 
    batch_size=32, 
    class_mode='categorical', 
    subset='validation'
)

print("Detected Classes:", train_generator.class_indices)
```

```python
# 4. Build the Convolutional Neural Network
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    layers.MaxPooling2D(2, 2),
    
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D(2, 2),
    
    layers.Flatten(),
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(train_generator.num_classes, activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()
```

```python
# 5. Train the Model
epochs = 10

history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=epochs
)
```

```python
# 6. Save and Download Model back to your computer
model.save('disease_cnn_model.h5')

from google.colab import files
files.download('disease_cnn_model.h5')
print("Model saved to disease_cnn_model.h5 and downloading... (Please place this in your d:/mini project/models/ folder!)")
```
